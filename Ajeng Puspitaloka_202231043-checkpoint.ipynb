```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Memuat gambar
image = cv2.imread('Ajeng.jpg')
```

```python
# Function to display images
def display_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i in range(len(images)):
        plt.subplot(2, 3, i+1)
        plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
# Gambar original
original = image.copy()

# Memutar Gambar
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))

# Ubah ukuran gambar (memanjang secara vertikal)
new_height = int(h * 1.5)  # Increase height by 50%
new_width = w  #Menyimpan lebarnya tetap sama
resized = cv2.resize(image, (new_width, new_height))

# Memotong gambar
cropped = image[50:200, 50:200]

# Membalik gambar
flipped = cv2.flip(image, 1)

# Menerjemahkan gambar
M = np.float32([[1, 0, 50], [0, 1, 100]])
translated = cv2.warpAffine(image, M, (w, h))

# Display all images
images = [original, rotated, resized, cropped, flipped, translated]
titles = ['Citra Asli', 'After Rotated', 'After Resized', 'After Cropped', 'After Flipped', 'After Translated']
display_images(images, titles)
```
